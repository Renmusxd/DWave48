```python
import numpy
from matplotlib import pyplot
```

$\mathbf{P} = \left(\frac{\beta_1}{\beta_2}\right)^{n_2 - n_1} \approx \left(\frac{\beta_1}{\beta_2}\right)^{-E_0(\beta_2 - \beta_1)}$

Find steps with constant $\mathbf{P}$...

```python
n = 10
ends = [(1.0, 10.0)]
ps = []

for _ in range(n):
    new_ends = []
    pps = []
    for start, stop in ends:
        betas = numpy.linspace(start,stop,10000)
        g1 = (betas - start)*numpy.log(start / betas)
        g2 = (stop - betas)*numpy.log(betas/stop)
        diffs = abs(g1 - g2)
        min_diffs = numpy.argmin(diffs)
        pps.append(numpy.exp(g1[min_diffs]))
        pps.append(numpy.exp(g2[min_diffs]))
        new_ends.append((start, betas[min_diffs]))
        new_ends.append((betas[min_diffs], stop))
    ps.append(pps)
    ends = new_ends
xs, ys = zip(*ends)
xs = numpy.asarray(xs)
ys = numpy.asarray(ys)
```

```python
pyplot.plot(ps[-1])
pyplot.yscale('log')
pyplot.show()
```

```python
pyplot.plot(ys - xs, 'x')
pyplot.ylabel('end - start')
pyplot.show()
```
