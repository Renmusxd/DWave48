```python
import numpy
from matplotlib import pyplot
```

```python
def sigzsigz(i,j, nvars=4):
    matrices = [numpy.eye(2)]*nvars
    matrices[i] = numpy.asarray([[1,0],[0,-1]])
    matrices[j] = numpy.asarray([[1,0],[0,-1]])
    return make_h(matrices)

def sigx(i, nvars=4):
    matrices = [numpy.eye(2)]*nvars
    matrices[i] = numpy.asarray([[0,1],[1,0]])
    return make_h(matrices)

def make_h(matrices):
    acc = matrices.pop(0)
    while len(matrices) > 0:
        acc = numpy.kron(matrices.pop(0), acc)
    return acc
```

```python
def ei(theta):
    return numpy.exp(1.0j*theta)
    
def make_ham(kx,ky, gamma=1.0):
    j_mat = numpy.asarray([
        [0,       1,       -ei(ky), -1],
        [1,       0,       -1,      -ei(kx)],
        [-ei(-ky), -1,      0,      -1],
        [-1,      -ei(-kx), -1,      0],
    ])
    h = numpy.zeros((16, 16), dtype=j_mat.dtype)
    for i in range(4):
        for j in range(i, 4):
            h += sigzsigz(i,j)*j_mat[i,j]
        h += gamma*sigx(i)
    return h

@numpy.vectorize
def eigenvalues(kx, ky, gamma=1.0):
    h = make_ham(kx, ky, gamma=gamma)
    return numpy.min(numpy.linalg.eigvals(h))
```

```python
kx = numpy.linspace(-numpy.pi, numpy.pi, 40)
ky = kx
KX, KY = numpy.meshgrid(kx, ky)
lams = eigenvalues(KX, KY)

maxlam = numpy.max(lams)
minlam = numpy.min(lams)

pyplot.contourf(KX, KY, lams[:,:], levels=numpy.linspace(minlam, maxlam, 20))
pyplot.colorbar()
pyplot.show()
```

```python
spins_fact = make_h([[2,1],[3,1],[5,1],[7,1]])
spins = numpy.asarray([spins_fact % 2 == 0,
         spins_fact % 3 == 0,
         spins_fact % 5 == 0,
         spins_fact % 7 == 0])*2 - 1

for kx in [0, numpy.pi]:
    for ky in [0, numpy.pi]:
        h = make_ham(kx,ky)
        l, v = numpy.linalg.eig(h)
        minv = v[numpy.argmin(l),:]
        print(numpy.sum(spins * numpy.expand_dims(numpy.abs(minv)**2, axis=0), axis=-1))
```

```python
from scipy import sparse
import scipy.sparse.linalg
```

```python
def make_ham_and_spin(nvars, edges, transverse):
    ham = sparse.lil_matrix((2**nvars,2**nvars), dtype=numpy.complex128)
    for i in range(2**nvars):
        state = [(i >> j) & 1 for j in range(nvars)]
        h = 0.0
        for (vara, varb), j in edges:
            if state[vara] == state[varb]:
                h += j
            else:
                h -= j

        ham[i,i] += h
        if transverse != 0:
            for k in range(nvars):
                j = i ^ (1 << k)
                if j > i:
                    ham[i,j] = ham[i,j] + transverse
                    ham[j,i] = ham[j,i] + transverse
#             for j in range(i+1, 2**nvars):
#                 b_state = [(j >> k) & 1 for k in range(nvars)]
#                 diffstate = [a^b for a,b in zip(state, b_state)]
#                 if sum(diffstate) != 1:
#                     continue
#                 for vark,s in enumerate(diffstate):
#                     if not s:
#                         continue
#                     ham[i, j] = ham[i,j] + transverse
#                     ham[j, i] = ham[j,i] + transverse

    spin_op = sparse.lil_matrix((2**nvars,2**nvars))
    for i in range(2**nvars):
        state = [(i >> j) & 1 for j in range(nvars)]
        spin_op[i,i] = numpy.sum(numpy.asarray(state)*2 - 1)**2
    return ham, spin_op
```

```python
kx, ky = 0, 0
# edges = [((0,1), 1.0), ((1,2), -1.0), ((2,3), -1.0), ((3,0), -1.0),
#          ((0,2), ei(ky)), ((1,3), ei(kx))]

x = numpy.arange(2)
y = numpy.arange(2)
n = numpy.arange(4)
X, Y, N = numpy.meshgrid(x, y, n)

INDX = Y*len(x)*len(n) + X*len(n) + N
SQR_BONDS = Y*len(x)*len(n) + X*len(n) + (N + 1)%len(n)

edges = [((a, b), 1 if a%len(n) == 0 and b%len(n) == 1 else -1) 
         for a,b in zip(INDX.flatten(), SQR_BONDS.flatten())]

for (sx, sy, sn) in zip(X.flatten(), Y.flatten(), N.flatten()):
    if sn == 1:
        a = sy*len(x)*len(n) + sx*len(n) + 1
        b = sy*len(x)*len(n) + ((sx+1)%len(x))*len(n) + 3
        edges.append(((a,b), -1))
    if sn == 2:
        a = sy*len(x)*len(n) + sx*len(n) + 0
        b = ((sy+1)%len(y))*len(x)*len(n) + sx*len(n) + 2
        edges.append(((a,b), -1))
```
